## Eksik Veri

* Eksik verilerin olduu zellikler

```python
df.columns[df.isnull().any()]
```

* Eksik veri olan kolonlarda toplam eksik gzlem says

```python
df.isnull().sum()
```

* Veri seti hikayesinden biliniyor ki "Bullet Type" zellii saysal olarak girilse de kategorik deikendir.Bunun hikaye ve deerlerden yola karak incelenmesi gerekmektedir.

* Range ,Bullet zelliklerin doldurulmasnda medyan ya da ortalama alnabilir. Bunun ile ilgili inceleme yapmak gerekir.

* BDMG ve HDMG zelliklerini ise inceleme yaplacak zellikler olmadndan 0 deeri ile doldurulabilir.

* BDMG ve HDMG zelliini doldurma

```python
for i in df.columns:
    if ("BDMG" in i) | ("HDMG" in i):
        df[i].fillna(0, inplace=True)
df.isnull().sum()
```

* Range zelliini doldurmak adna incelemek

```python
df.Range.mean()
```

```python
df.Range.median()
```

* Range zellii iin ortalama ve median yakn olduundan her ikisi de kullanlabilir. Ortalama deeri tercih edelim.

```python
df.Range.fillna(df.Range.mean(), inplace=True)
df.isnull().sum()
```

* Bullet_Speed zellii iin ayn inceleme yaplr.

```python
df.Bullet_Speed.mean()
```

```python
df.Bullet_Speed.median()
```

* Buradaki fark silah trleri ile ilikili olabilir. O yzden gruplama yapp ayn incelemeyi yapmak mantkl olabilir.

```python
df.groupby("Weapon_Type").aggregate({"Bullet_Speed": [np.mean, np.median]})
```

* Grupland zaman ortalama ve medyan deerlerinin daha yaknlatn gzlemlenebilir. Fakat baz silahlarda fark azken bazlarnda fazla olduundan silah trne gre mermi hznn medyan ile doldurmak daha iyi olacaktr.

```python
gruplanan_hiz_sozlugu =dict(df.groupby("Weapon_Type")["Bullet_Speed"].median())
for x,y in gruplanan_hiz_sozlugu.items():
    if math.isnan(float(y)) :
        gruplanan_hiz_sozlugu[x] = 0.0
        
df_bullet = df[df["Bullet_Speed"].isnull().values]
indeks = df_bullet.index.tolist()
deger = []

for i in df_bullet["Weapon_Type"] :
    deger.append(gruplanan_hiz_sozlugu[i])

df_bullet.drop("Bullet_Speed" , axis=1)
df_bullet["Bullet_Speed"] = deger
df[df["Bullet_Speed"].isnull().values] = df_bullet

df.isnull().sum()
```

* Bullet Type zellii eksik veri incelemesi

```python
df["Bullet_Type"].unique()
```

Burada eksik olan deerler bo braklmam yakn dv silahlarnn mermisi olmadndan mermi tr belirtilmeyip nan braklm. Bu verilerin byle braklmas daha doru olacaktr.nk snrl veriye sahip veri setinde tek zellik iin gzlem silmek mantkl deil hatta bilinli olarak yani o silahlarn bu zellii olmadndan bo braklmasna ramen ilem yaplmas doru olmayacaktr.

## Aykr Deer Tespiti

* Bu veri setinde gzlem birimi ok olmadndan aykr gzlemlerle denk gelindiinde silmek yerine basklama yntemi kullanlmaldr.
* Aykr gzlem incelemesi yaplmas gereken zellikler; Damage, Damage_Per_Second, Magazine_Capacity, Range, Bullet_Speed, Rate_of_Fire.
* Aykr gzlemleri grmek adna boxplot iyi bir aratr.

```python
def box_plot(df , f1, f2):
        sns.boxplot(x=f1 , y=f2 , data = df)
        plt.show();
```

```python
x_ekseni = ["Damage","Damage_Per_Second", "Magazine_Capacity", "Range", "Bullet_Speed", "Rate_of_Fire"]
y_ekseni = "Weapon_Type"
for i in x_ekseni:
    box_plot(df,i,y_ekseni)
```

* Silah trlerine gre eitli aykrlklar sz konusu fakat bunlarn zerine herhangi bir ilem (silmek ya da basklama gibi) uygulanmayacaktr.
